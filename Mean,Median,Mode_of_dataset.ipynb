```python
import pandas as pd
df = pd.read_csv("result.csv")
df = df[df['country_or_area']=="Africa +"]
df = df[df['category']=="bananas"]
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title('Production/Year')
plt.xlabel('Year')
plt.ylabel('Value(in Ha)')
plt.plot(df.year, df.production_value, label='Value')
plt.legend()
```

```python
#making a new csv file
new_df = df[['year', 'crops_value']]
convert_dict = {'year': int}
new_df = new_df.astype(convert_dict)
new_df.to_csv('Target.csv')  
```

```python
#making a column with some missing values
import numpy as np
df = new_df.head(8)
numbers = [1,2,5,np.nan,6,np.nan,np.nan,8]
df['Numbers']=numbers
df
```

```python
pf = df
df.fillna(df.mean())
```

```python
tf = pf
pf.fillna(df.median())
```

```python
df['Numbers'] = tf['Numbers'].fillna(df['Numbers'].mode()[0])
df
```
